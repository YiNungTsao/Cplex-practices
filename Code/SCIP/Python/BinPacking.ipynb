```python
import pyscipopt
```

```python
B = 100

items = 11
num_sets = 11
s = {}
for i in range(items):
    s[i] = 0
s[0] = 48
s[1] = 30
s[2] = 19
s[3] = 36
s[4] = 36
s[5] = 27
s[6] = 42
s[7] = 42
s[8] = 36
s[9] = 24
s[10] = 30
```

```python
model = pyscipopt.Model("BinPacking")

# Create variables
x = {}
for i in range(items):
    for j in range(num_sets):
        x[i,j] = model.addVar(lb=0, ub=1, vtype="B", name="x(%s,%s)"%(i,j))

y = {}
for j in range(num_sets):
    y[j] = model.addVar(lb=0, ub=1, vtype="B", name="y(%s)"%(j))
        
        
# Create objective function
model.setObjective(pyscipopt.quicksum(y[j] for j in range(num_sets)), sense="minimize")
```

```python
# Create constraints
# Capacity constraints
for j in range(num_sets):
    model.addCons(pyscipopt.quicksum(s[i]*x[i,j] for i in range(items)) <= B*y[j])

# Assignment constrains
for i in range(items):
    model.addCons(pyscipopt.quicksum(x[i,j] for j in range(num_sets)) == 1)
```

```python
model.writeProblem("BinPacking.lp")
model.optimize()
status = model.getStatus()
print(f"Solution status: {status}")
if status == "optimal":
    print(f"Objective value = {model.getObjVal()}")
    for i in range(items):
        for j in range(num_sets):
            print(f"{x[i,j]} = {model.getVal(x[i,j])}")
    for j in range(num_sets):
        print(f"{y[j]} = {model.getVal(y[j])}")
```
